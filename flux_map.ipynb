```python
import numpy as np
import matplotlib.pyplot as plt

alpha = np.linspace(0, 2*np.pi, 500)
theta = np.linspace(-np.pi/2, np.pi/2, 500)
A, T = np.meshgrid(alpha, theta)

cases = [(0, 1), (1, 1), (2, 1)]
prefactor = np.sin(T)**2 * np.cos(A)**2 + np.sin(A)**2

# Precompute all Z to get global color scale
Zs = []
for N, Omega in cases:
    Z1 = prefactor * np.real(np.sqrt((1 - (N/Omega)**2) * (4*np.sin(T)**2 - 1) + 4*np.cos(T)**2 * np.sin(A)**2 + 0j))
    Z2 = prefactor * np.real(np.sqrt((1 - (N/Omega)**2) * (4*np.sin(T)**2 - 1) + 0j))
    Zs.append((Z1, Z2))

vmin = 0
vmax = max(Z.max() for pair in Zs for Z in pair)

fig, axes = plt.subplots(3, 2, figsize=(12, 10), sharey=True, sharex=True)

for row, ((N, Omega), (Z1, Z2)) in enumerate(zip(cases, Zs)):
    for col, Z in enumerate([Z1, Z2]):
        ax = axes[row, col]
        im = ax.pcolormesh(A, T, Z, cmap="magma", shading="gouraud", vmin=vmin, vmax=vmax)
        ax.axhline( np.pi/6, color="white", linewidth=0.8, linestyle="--")
        ax.axhline(-np.pi/6, color="white", linewidth=0.8, linestyle="--")
        ax.set_yticks([-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2])
        ax.set_yticklabels([r"$-\pi/2$", r"$-\pi/4$", r"$0$", r"$\pi/4$", r"$\pi/2$"])
        ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
        ax.set_xticklabels([r"$0$", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
        if row == 2:
            ax.set_xlabel(r"$\alpha$")
        if col == 0:
            ax.set_ylabel(r"$\vartheta$")
        ax.set_title(rf"$N/\Omega = {N}/{Omega}$")

fig.colorbar(im, ax=axes, location="right", shrink=0.8)
plt.savefig("flux_map_comparison.pdf", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = np.linspace(0, 2*np.pi, 200)
theta = np.linspace(-np.pi/2, np.pi/2, 200)
A, T = np.meshgrid(alpha, theta)

cases = [(0, 1), (1, 1), (2, 1)]
prefactor = np.sin(T)**2 * np.cos(A)**2 + np.sin(A)**2

trapz = np.trapezoid if hasattr(np, "trapezoid") else np.trapz

Zs, profiles = [], []
for N, Omega in cases:
    Z1 = prefactor * np.real(np.sqrt((1 - (N/Omega)**2) * (4*np.sin(T)**2 - 1) + 4*np.cos(T)**2 * np.sin(A)**2 + 0j))
    Z2 = prefactor * np.real(np.sqrt((1 - (N/Omega)**2) * (4*np.sin(T)**2 - 1) + 0j))
    Zs.append((Z1, Z2))
    profiles.append((trapz(Z1, alpha, axis=1), trapz(Z2, alpha, axis=1)))

vmin = 0
vmax = max(Z.max() for pair in Zs      for Z in pair)
imax = max(I.max() for pair in profiles for I in pair)

col_titles = ["Exact", "Traditional Approx.", r"$\int d\alpha$"]

fig, axes = plt.subplots(
    3, 3, figsize=(9, 7),
    gridspec_kw={"width_ratios": [1, 1, 0.55]},
    constrained_layout=True,
)

yticks      = [-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2]
yticklabels = [r"$-\pi/2$", r"$-\pi/4$", r"$0$", r"$\pi/4$", r"$\pi/2$"]
xticks      = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
xticklabels = [r"$0$", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"]

for row, ((N, Omega), (Z1, Z2), (I1, I2)) in enumerate(zip(cases, Zs, profiles)):
    last_row = (row == 2)

    for col, Z in enumerate([Z1, Z2]):
        ax = axes[row, col]
        im = ax.pcolormesh(A, T, Z, cmap="magma", shading="gouraud", vmin=vmin, vmax=vmax)
        ax.axhline( np.pi/6, color="white", linewidth=0.8, linestyle="--")
        ax.axhline(-np.pi/6, color="white", linewidth=0.8, linestyle="--")
        ax.set_xticks(xticks)
        ax.set_yticks(yticks)
        ax.set_xticklabels(xticklabels if last_row else [])
        ax.set_yticklabels(yticklabels if col == 0 else [])
        if last_row:
            ax.set_xlabel(r"$\alpha$")
        if col == 0:
            ax.set_ylabel(r"$\vartheta$")

    ax3 = axes[row, 2]
    ax3.plot(I1, theta, color="mediumpurple", label="Exact")
    ax3.plot(I2, theta, color="tomato", linestyle="--", label="Traditional Approx.")
    ax3.axhline( np.pi/6, color="gray", linewidth=0.8, linestyle="--")
    ax3.axhline(-np.pi/6, color="gray", linewidth=0.8, linestyle="--")
    ax3.set_yticks(yticks)
    ax3.set_yticklabels([])
    ax3.set_xlim(0, imax)
    ax3.set_ylim(theta[0], theta[-1])

for col, title in enumerate(col_titles):
    axes[0, col].set_title(title)

for row, (N, Omega) in enumerate(cases):
    axes[row, 0].annotate(
        rf"$N/\Omega = {N}$",
        xy=(0, 0.5), xycoords="axes fraction",
        xytext=(-0.25, 0.5), textcoords="axes fraction",
        ha="center", va="center", rotation=90, fontsize=9,
    )

axes[0, 2].legend(fontsize=8, loc="best")

fig.colorbar(im, ax=axes[:, :2], location="bottom", shrink=0.6, pad=0.02, aspect=40)

plt.savefig("flux_map_comparison_alphaintegral.pdf", bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator
import cartopy.crs as ccrs
import cartopy.feature as cfeature

alpha = np.linspace(0, 2*np.pi, 200)
theta = np.linspace(-np.pi/2, np.pi/2, 200)
A, T = np.meshgrid(alpha, theta)

cases = [(0, 1), (1, 1), (2, 1)]
prefactor = np.sin(T)**2 * np.cos(A)**2 + np.sin(A)**2
trapz = np.trapezoid if hasattr(np, "trapezoid") else np.trapz

profiles = []
for N, Omega in cases:
    Z1 = prefactor * np.real(np.sqrt((1 - (N/Omega)**2) * (4*np.sin(T)**2 - 1) + 4*np.cos(T)**2 * np.sin(A)**2 + 0j))
    Z2 = prefactor * np.real(np.sqrt((1 - (N/Omega)**2) * (4*np.sin(T)**2 - 1) + 0j))
    profiles.append((trapz(Z1, alpha, axis=1), trapz(Z2, alpha, axis=1)))

vmax = max(I.max() for pair in profiles for I in pair)

lon_deg = np.linspace(-180, 180, 200)
lat_deg = np.degrees(theta)
LON_deg, LAT_deg = np.meshgrid(lon_deg, lat_deg)

fig, axes = plt.subplots(
    2, 3, figsize=(12, 6),
    subplot_kw={"projection": ccrs.Mollweide()},
    constrained_layout=True,
)

for col, ((N, Omega), (I1, I2)) in enumerate(zip(cases, profiles)):
    for row, I in enumerate([I1, I2]):
        ax = axes[row, col]
        Z_map = np.outer(I, np.ones(len(lon_deg)))
        im = ax.pcolormesh(LON_deg, LAT_deg, Z_map, cmap="magma", shading="gouraud",
                           vmin=0, vmax=vmax, transform=ccrs.PlateCarree())
        ax.add_feature(cfeature.COASTLINE, linewidth=0.4, edgecolor="white")
        gl = ax.gridlines(linewidth=0.3, color="white", alpha=0.5, linestyle=":")
        gl.xlocator = FixedLocator([-120, -60, 0, 60, 120])
        gl.ylocator = FixedLocator([-60, -30, 0, 30, 60])
        for lat in [30, -30]:
            ax.plot([-180, 180], [lat, lat], transform=ccrs.PlateCarree(),
                    color="white", linewidth=0.8, linestyle="--")
    axes[0, col].set_title(rf"$N/\Omega = {N}$")

row_labels = ["Exact", "Traditional Approx."]
for row, label in enumerate(row_labels):
    axes[row, 0].text(-0.05, 0.5, label,
                      transform=axes[row, 0].transAxes,
                      ha="right", va="center", rotation=90, fontsize=9)

fig.colorbar(im, ax=axes, location="bottom", shrink=0.5, pad=0.02, aspect=40)

plt.savefig("flux_map_mollweide.pdf", bbox_inches="tight")
plt.show()
```
