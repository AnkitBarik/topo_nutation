```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_array = np.array([1.001, 1.01, 1.1])
H = np.linspace(1,100e3,500)
```

```python
k0 = 4*np.pi/H
nu = 0.7
Omega = 2*np.pi/86400
```

```python
hrms = np.zeros((len(N_array), len(H)))
for i, N in enumerate (N_array):
    hrms[i] = np.sqrt((nu-0.5)/np.sqrt(N**2-1)/k0/Omega)
```

```python
plt.rc('text', usetex=True)
for i in range(hrms.shape[0]):
    plt.plot(H/1e3, hrms[i]/1e3, label=N_array[i])
plt.ylabel(r'$h_\mathrm{rms} [\mathrm{km}]$', fontsize=14)
plt.xlabel(r'$H [\mathrm{km}]$', fontsize=14)
plt.grid()
plt.legend()
plt.savefig("Diss.pdf")
plt.show()
```
