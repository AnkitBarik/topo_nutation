```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fixed parameters
Omega = 2*np.pi/86400
R = 3486e3
rho = 13e3
U0 = 8.7e-5
```

```python
N_array = np.array([1.1, 1.3, 2]) # B.V. frequency
H_array = np.array([0.6,0.7,0.8]) # Hurst exponent
d_array = np.array([50,100,200]) # depth
```

```python
def hrms(N,H,d,target):
    lambda0 = np.linspace(1,0.5*d,500)
    k0 = 2*np.pi/lambda0
    return lambda0, np.sqrt( (2**(7+4*H) * d * (d*k0)**(-2*H) * np.pi**(2*H-1) * (2*H-1)) / (17 * R**2 * U0**2 * H * rho * Omega * np.sqrt(N**2-1)) ) * np.sqrt(target)
```

```python
plt.rc('text', usetex=True)
fig, axes = plt.subplots(3, 2, figsize=(8, 10), sharex=True, sharey=True)  # Adjust figsize as needed

axes[0,0].set_ylim(2, 15)
axes[0,0].set_xlim(-10, 400)

for k, d in enumerate(d_array[1:]):
    axes[-1,k].set_xlabel(r'$d [\mathrm{km}]$', fontsize=12)
    for j, H in enumerate(H_array):
        axes[j,0].set_ylabel(r'$h_\mathrm{rms} [\mathrm{km}]$', fontsize=12)
        ax = axes[j,k]  # Select the current subplot
        for i, N in enumerate(N_array):
            lambda0, h = hrms(N, H, d, 9.5e6)
            lambda0, hplus = hrms(N, H, d, 9.5e6+1.4e6)
            lambda0, hminus = hrms(N, H, d, 9.5e6-1.1e6)
            ax.plot(lambda0/1e3, h/1e3, label=fr'$N/\Omega = {N_array[i]}$')
            ax.fill_between(lambda0/1e3, hminus/1e3, hplus/1e3, alpha=0.2)
            ax.fill_betweenx(lambda0, 0.5*d/1e3,400, color='gray', alpha=0.1)
            
        ax.grid()
        
        # Add a title to each subplot
        ax.set_title(fr'$\mathcal{{H}} = {H}, d = {d_array[k+1]/1e3}~\mathrm{{km}}$', fontsize=14)

axes[0,0].legend()

# Adjust layout
plt.tight_layout()

#plt.savefig("constraint-col.pdf")
plt.show()
```
